```python
import pandas as pd

data = {
    'gender': ['female', 'male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'male'],
    'race_ethnicity': ['group A', 'group B', 'group A', 'group B', 'group C', 'group C', 'group D', 'group D', 'group E', 'group E'],
    'parental_education': ['bachelor', 'some college', 'master', 'associate', 'high school', 'some college', 'associate', 'bachelor', 'high school', 'master'],
    'lunch': ['standard', 'free/reduced', 'standard', 'free/reduced', 'standard', 'standard', 'free/reduced', 'standard', 'free/reduced', 'standard'],
    'test_preparation_course': ['none', 'completed', 'none', 'completed', 'none', 'none', 'completed', 'none', 'completed', 'completed'],
    'math_score': [72, 69, 90, 47, 76, 71, 88, 40, 64, 95],
    'reading_score': [72, 90, 95, 57, 78, 83, 95, 43, 60, 98],
    'writing_score': [74, 88, 93, 44, 75, 78, 92, 39, 65, 99]
}

df = pd.DataFrame(data)
```

```python
# Use .head(), .info(), .describe() to get a quick overview 
df.head(10)
df.info()
df.describe()
```

```python
# âœ… 2. Data Profiling
# What are the data types of each column?
df.dtypes
# How many missing values are there?
df.isnull().sum()
# What are the unique values for gender, lunch, etc.?
df.nunique()
# What is the skewness and kurtosis of math_score, reading_score, and writing_score?
print(df["math_score"].skew())
print(df["reading_score"].skew())
print(df["writing_score"].skew())
# negative skew mean left skew meaning majority socres high most student scored above average
```

```python
# âœ… 3. Handling Missing & Duplicates

# Are there any missing values? If yes, how to deal with them?
df.isnull()
# Any duplicate rows?
```

```python
# âœ… 4. Univariate Analysis
# Plot histograms of math_score, reading_score, writing_score
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df["math_score"],bins=10,kde=False,color="red")
plt.show()
sns.histplot(df["reading_score"],bins=10,kde=False,color="green")
plt.show()
sns.histplot(df["writing_score"],bins=10,kde=False,color="yellow")
plt.show()
# Plot KDE plots for the same columns
sns.kdeplot(df["math_score"])
plt.show()
sns.kdeplot(df["reading_score"])
plt.show()
sns.kdeplot(df["writing_score"])
plt.show()

# Create box plots to detect outliers in scores
sns.boxplot(df["math_score"])
plt.show()
# this box plot means minimum score is 40 maximum score is 95 meadian score is around 70 and 25 percent of student scored below 65 75 percent scored below 85 is this understanging correct


sns.boxplot(df["reading_score"])
plt.show()
sns.boxplot(df["writing_score"])
plt.show()

# Plot countplots for gender, lunch, and test_preparation_course
sns.countplot(df["gender"],color="pink")
plt.show()
sns.countplot(df["lunch"],color="blue")
plt.show()
sns.countplot(df["test_preparation_course"],color="yellow")
plt.show()
```

```python
# âœ… 5. Bivariate Analysis
# Correlation matrix between the 3 scores
corr_score=df[["math_score","reading_score","writing_score"]].corr()
sns.heatmap(corr_score,annot=True,cmap="coolwarm")
plt.show()
# Use a box plot: test_preparation_course vs math_score
sns.boxplot(x=df["test_preparation_course"],y=df["math_score"],data=df)
plt.show()
# Use a violin plot: lunch vs reading_score
# Use a swarm/strip plot: gender vs writing_score
```
