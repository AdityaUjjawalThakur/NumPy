```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("train.csv")
# Load the dataset and display the first 10 rows.
print(df.head(10))
# Show number of passengers (rows) and features (columns).
print(f"Number of passengers{df.shape[0]}")
print(f"Number of features{df.shape[1]}")
# Check for missing values in each column.
df.isnull().sum()
# Drop the Cabin column temporarily.
# df.drop(columns="Cabin")
# Find the average age of passengers grouped by Sex.
avg_age=df.groupby("Sex")["Age"].mean()
print(avg_age)
# Count how many passengers survived and didnâ€™t survive.
df.groupby("Survived")["PassengerId"].count()
# df["Survived"].value_counts()
# Find the survival rate grouped by Pclass (passenger class).
print(df.groupby("Pclass")["Survived"].mean())
#  this statement means aa the data is grouped according to pclass then we are doing calculation on survivrd column adnof each group we are taking mean
# Get the names of all passengers under age 18 who survived.
df[(df["Age"]<18)&(df["Survived"]==1)]["Name"]
# Create a new column FamilySize = SibSp + Parch.
df["Familysize"]=df["SibSp"]+df["Parch"]
print(df.head(10))
# Group by Sex and Pclass, then calculate average fare and survival rate.
df.groupby(["Sex","Pclass"])[["Fare","Survived"]].mean()
# Sort the dataset by Fare in descending order and display the top 5.
df.sort_values(by="Fare",ascending=False).head(5)
```
