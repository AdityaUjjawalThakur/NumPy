```python
import pandas as pd
#series
#from a list
slaries=[10000,20000,30000,40000]
s=pd.Series(slaries)
print(s)
#custom indexing
print(pd.Series(slaries,index=["jan","feb","mar","april"]))
#from a dictonary
dic={"name":"Aditya ujjawal","age":22,"dept":"cse","marks":[96,89.78]}
sdic=pd.Series(dic)
print(sdic)
#Attributes
print(sdic.values)
print(sdic.index)
print(sdic.dtype)


```

```python
import numpy as np
import pandas as pd
# Dataframe is basically a 2d stucture to represent a table like structure in which colum repesent a series 

#from list of dictionaries
df=pd.DataFrame([
    {"name":"Aditya","age":25,"marks":89},
    {"name":"Abhinav ","age":19,"marks":90},
    {"name":"Yuvraj ","age":19,"marks":20},
    {"name":"Utkarsh ","age":10,"marks":60},
    {"name":"Mrinal ","age":40,"marks":70}

])
#each row represent a record 
# print(df)
#from dictionary of list
df2=pd.DataFrame({
    "names":["Aditya","Abhinav","Yuvraj"],
    "age":[20,19,19],
    "marks":[89,90,60]
})
# print(df2)
#numpy arrays with name
a=np.array([["aditya",20,89],["abhinav",19,90],["yuvraj",21,45]])
df3=pd.DataFrame(a,columns=["names","age","marks"])
print(df3)
print(df.index)
print(df.columns)
print(df.info()) #it gives info about no oh rows ,column names,non null counts,data types memory usages
print(df.describe())#give summary about columns count,mean,std,minmnax,25 ,50,75 percitle data like these
```

```python
#selecting and filtering data in daraframe

print(df[["name","age"]])
print(df.loc[0])
print(df.loc[:,"name"])# first argument is about row and 2nd argument is about column
print(df.loc[:,:])# all rows and each cell
print(df.loc[1,["name"]])# data from row 1 and column name it picks that cell
#iloc
print(df.iloc[0]) # in iloc we can access only by position that is through index while in loc we
# can have both index position and as well as positions
# filtering rows in dataframe
print(df["age"]>20) #df["age"]>20 gives boolean values and puting that into actual dataframe return row that is true
print(df[df["age"]>20])
print(df[(df["age"]>10) & (df["marks"]>80)])
# modifying columns in a dataframe
#add a column
df["passed"]=df["marks"]>80
print(df)
df["college"]=["Nmamit","VTU","Govind","RNSIT","NITT"]
print(df)
#deleting columns
df.drop("college",axis=1)
# #deleting a row
df.drop(2,axis=0,inplace=True)
print(df)
```
