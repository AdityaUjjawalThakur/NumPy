```python
#Groupby and aggregation
#Group by basically very simalar to sql group by it groups the data on some column or categories 
#and can perform some operation obver the group
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "Department": ["CS", "CS", "EE", "EE", "ME", "CS", "ME"],
    "Student": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace"],
    "Marks": [85, 90, 78, 88, 76, 92, 89]
}
df=pd.DataFrame(data)
print(df)
print(df.groupby("Department")["Marks"].mean())
#Number of student per department
stc=df.groupby("Department")["Student"].count()

print(stc)
# stcd=pd.DataFrame(stc)
# stcd.set_index(0)["Student"].plot(kind="bar") this is because stc is in series and when we convert into dataframe it does not retain its column name
stc.plot(kind="bar", ylabel="Number of Students", title="Students per Department")

```

```python
#Merge 
#The Difference Between Merge And Concat is concat just stack the tables along rows or columns while join combine tables 
import pandas as pd
student_df=pd.DataFrame({

    "usn":[100,101,102,103],
    "name":["Alice","Joe","Shogun","charlie"],
    "Department":["Cse","ise","cse","Me"]
})
attendence_df=pd.DataFrame(
{
    "usn":[100,101,102,103],
    "attendence":[89,90,96,76]
})
pd.merge(student_df,attendence_df,on="usn")
#for left and right join all row of corresponding left and right rows and matching rows of right table if rows is not matchinh then nan in that place
```

```python
#apply function
# this function helps to apply a function to each values of a column and rows 
#add a column to attendence_df if attenence is below 80 noit elogible for exams
def iseligible(e):
    if e<80:
        return "No"
    else:
        return "Yes"
attendence_df["eligible"]=attendence_df["attendence"].apply(iseligible)
print(attendence_df)
```
