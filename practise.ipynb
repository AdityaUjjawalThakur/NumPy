```python
# 1. Basic DataFrame Exploration
import numpy as np
import pandas as pd
df=pd.read_csv('StudentsPerformance.csv')

# Load the dataset and display the first 10 rows.
print(df.head(10))
# Check the number of rows and columns.
# print(df.describe())
# print(f"No Of Rows Are{df.shape[0]}")
# print(f"No of column are {df.shape[1]}")
# Get column names and data types.
# print(df.columns)
# print(df.dtypes)
# Find missing values in the dataset (if any).
print(df.isnull().values.any()) # return true if any missing values
print(df.isnull())# it return a table of true and false
print(df.isnull().sum())# it returnn how many valuues in column is null

```

```python
# 2. Selection and Filtering
# Select only the gender and math score columns.
# print(df[["gender","math score"]])
#Filter rows where math score is greater than 90
# df[df["math score"]>90]
# Get the records of female students who scored above 80 in all three subjects.
df[(df["gender"] == "female") &
   (df["math score"] > 80) &
   (df["reading score"] > 80) &
   (df["writing score"] > 80)]
```

```python
# 3. Data Modification
#Add a new column average_score that’s the mean of the three exam scores.
# df["avg_score"]=df[["math score","reading score","writing score"]].mean(axis=1)# by default bmean operates column wise axis=0
# print(df.columns)
# Update the test preparation course column to be "completed" or "not completed" in title case.
# df[df["test preparation course"]=="none"]="Not completed" wrong
df["test preparation course"] = df["test preparation course"].replace({
    "none": "Not completed",
    "completed": "Completed"
})
print(df)


# Drop the column lunch from the dataset temporarily (without modifying the original DataFrame).
df.drop(columns="lunch")
```

```python
# Sort the dataset by reading score in descending order.Group by gender and find the average of each exam score.
import pandas as pd
df_sorted=df.sort_values(by="reading score",ascending=False)
df_sorted.groupby("gender")[["math score","reading score","writing score"]].mean()
```
