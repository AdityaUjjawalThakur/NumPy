```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "Name": ["Aditya", "Abhinav", "Yuvraj", "Utkarsh", "Mrinal"],
    "Math_Score": [88, 92, 75, 65, 70],
    "Science_Score": [90, 85, 80, 70, 60],
    "English_Score": [85, 78, 82, 60, 72],
    "Study_Hours": [12, 14, 10, 8, 9],
    "Attendance (%)": [95, 98, 90, 85, 80]
}

df = pd.DataFrame(data)
print(df)
df.set_index("Name",inplace=True)

df.plot()
plt.title("Student Scores across the the subject ")
plt.xlabel("Student")
plt.ylabel("scores")
plt.grid(True)
plt.show()
```

```python
#Line Graph
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
movie = {
    "Movie": ["RRR", "Pathaan", "KGF 2", "Jawan", "Drishyam 2"],
    "Rating": [8.5, 7.9, 8.2, 7.5, 8.1],
    "Collection": [1200, 1050, 950, 1100, 230],  # in crores
    "MusicScore": [8.0, 7.5, 8.7, 6.5, 7.0]
}

df=pd.DataFrame(movie)
print(df)
# #line graph
# # df.set_index("Movie")[["Rating","Collection","MusicScore"]].plot()
# # df.set_index("Movie")[["Rating", "Collection", "MusicScore"]].plot(kind="line", marker='o')
# df.set_index("Movie")[["Rating", "Collection", "MusicScore"]].plot(
#     kind="line", 
#     marker='o',
#     subplots=True,
#     layout=(3,1),  # 3 rows, 1 column
#     figsize=(8, 10),
#     sharex=True
# )
# plt.suptitle("Metrics per Movie (Separate Scales)")
# df.set_index("Movie",inplace=True) 
# we are getting plot but problem is collection has very big values as compared to rating and music score bcz of which it is not coming in graph clearly
# this is another way same problrm here 
# df.set_index("Movie",inplace=True)
# df[["Collection","Rating","MusicScore"]].plot()
#solution is we can have subplots comparing each categories per plot
ax=df.set_index("Movie")[["Rating","Collection","MusicScore"]].plot(
    subplots=True,
    kind="line",
    marker='o'
)
plt.xlabel("Movies Names")
#to label y axis 
ax[0].set_ylabel("Rating")
ax[1].set_ylabel("Collection")
ax[2].set_ylabel("Music Score")
df.plot()





# df.plot()
```

```python
#Box Plot
#observe the box bottom of the box here in image around 65 tells that 25 percent of student scored below 65
# upper line in case of maths seems 88 which means 75 percent of student scored below 88 and median marks is represented by green
import pandas as pd
import matplotlib.pyplot as plt
data={
    "Maths":[75, 88, 95, 60, 70, 85, 100, 55, 65, 80],
    "English":[78, 85, 92, 58, 66, 89, 99, 50, 68, 72],
    "Science":[72, 84, 91, 62, 69, 87, 98, 53, 67, 75]
}
df=pd.DataFrame(data)
df.plot(kind="box")
plt.grid(True)
```

```python
#pie chart 
#it is used to compare withing same category like how much student is in cse as compared tp all students in all brach combined
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "Department": ["CSE", "ECE", "Mechanical", "Civil"],
    "Students": [120, 100, 80, 60]
}
df=pd.DataFrame(data)
df.set_index("Department").plot(kind="pie",y="Students",autopct="%1.1f%%")
plt.title("Student Distribution by Department")
```

```python
#ScatterPlot Shows Correlation between Two Values Example numver of hours studied and mars scored
import matplotlib.pyplot as plt
data={
    "Hours_Studied": [1, 2, 3, 4, 5, 6, 7],
    "Scores": [35, 45, 50, 55, 65, 70, 85]
}
df=pd.DataFrame(data)
df.plot(kind="scatter",x="Hours_Studied",y="Scores",color="red")
plt.title("Scores Vs Study Hours")
```
