```python
import pandas as pd
import numpy as np
df=pd.read_csv("titanic.csv")
df.isnull().sum()
df.head(10)
```

```python
# # Empirical Rule (68â€“95â€“99.7 Rule)
# # Applies to normal distributions
# Standard Deviation means how data is distributed or spresd around mean like if mean age is 30 and
# most of ppl age is aronuf 25 28 like this then deviation is less if ahge is lik 4 89 67 very 
# distant from mean this means spread is more and stds is more
# emperical rile is imagine mean age of paenger i 30 and std is 10 years then 
# 68% of passenger is between(30+-10) mean between 20 and 40 year
# 95 % is is between 10 and 50 year(30+-20)
# \99.7% are between 0 and 60 years (30 Â± 30)
df["Age"].dropna().std()
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

ages = df["Age"].dropna()
mean = ages.mean()
std = ages.std()

# Visualize distribution
sns.histplot(ages, kde=True)

# Plot mean and SD lines
plt.axvline(mean, color='black', linestyle='--', label='Mean')
plt.axvline(mean + std, color='green', linestyle='--', label='Â±1 SD')
plt.axvline(mean - std, color='green', linestyle='--')
plt.axvline(mean + 2*std, color='orange', linestyle='--', label='Â±2 SD')
plt.axvline(mean - 2*std, color='orange', linestyle='--')
plt.axvline(mean + 3*std, color='red', linestyle='--', label='Â±3 SD')
plt.axvline(mean - 3*std, color='red', linestyle='--')
plt.legend()
plt.title("Empirical Rule Applied on Titanic Age Distribution")
plt.show()


```

```python
#zscore
from scipy.stats import zscore
# z=(X-mean)/standard deviation  
# z=0 the value is exaclt at themean
# z=+1 value is 1std above the mean
# z=-1 value is 1std below the mean
# z=+-2 value is 2 std above or below the mean
# z>+3 or z<-3 possible outlier
# Letâ€™s say in the Titanic dataset:
# Mean age = 30
# standard deviation = 10
# Then:
# A person aged 50 has Z = (50âˆ’30)/10=2 â†’ 2 SDs above mean
# A person aged 15 has Z = (15âˆ’30)/10=âˆ’1.5 â†’ 1.5 SDs below mean
ages=df["Age"].dropna()
z_scores=zscore(df["Age"].dropna())
print(z_scores)
ages[(z_scores>3)|(z_scores<-3)]


```

```python
#outlierDetection 
# zscore if zscore is more than 3 or less than -3 that that is on outlier
# Using IQR (Interquartile Range)
# iqr=q3-q1
# lower bound=q1-1.5*iqr
# upper bound=q3+1.5*iqr
# is value is out of these bound than it is an outlier
ages=df["Age"].dropna()
q1=ages.quantile(0.25)
q3=ages.quantile(0.75)
iqr=q3-q1
lowerbound=q1-1.5*iqr
upperbound=q3+1.5*iqr


print(q1)
print(q3)
print(iqr)
ages[(ages<lowerbound) |(ages>upperbound)] #outlier age
#visual methods
# box plot is best suited for neumerical analysis of outliers
# sctter plot is buest suited to find outliers in relationship between two neumerical values
```

```python
# 3. Sampling Distributions
# When analyzing a huge dataset or an entire population is difficult (due to size, cost, time, etc.), we can:
# Take a random sample of fixed size
# Calculate statistics like mean, median, etc.
# Repeat this process many times
# he distribution of those statistics is called a sampling distribution
# And thanks to the Central Limit Theorem, this sampling distribution:
# Becomes normally distributed (even if the original data is not)
# Allows us to estimate and make inferences about the population
# CENTRAL LIMIT THEOREM
# Even if the original data is not normally distributed,
# The distribution of sample means (i.e., the sampling distribution) will become normal,
# As long as:
# Samples are random and independent
# Sample size is reasonably large (n â‰¥ 30 is a good rule of thumb)
# A normal distribution lets us:
# Use the mean confidently (it's stable and predictable)
# Calculate probabilities easily (empirical rule, z-scores)
samples_means=[]
for _   in range(1000):
    sample=df["Fare"].dropna().sample(n=50,replace=True)
    samples_means.append(sample.mean())
sns.histplot(data=samples_means,kde=True,color="green")
plt.axvline(np.mean(samples_means),color="red",label="mean of sample")
plt.title("Sampling Distribution of the Mean (Fare)")
plt.show()
# How Clt helps
# inference statistics is all about making prediction about the population based on sample dataset thanks to clt which tells that if we do sampling 
# distribution n numner of times gives n is significant enough even if sample is normal distributed evenetual sample will becomes normal distributed 
# which makes the analysis relaible like mean becomes more accurate 
# Thanks to CLT, we can:
# Use small samples from large datasets
# Model probabilities and outcomes
# Run machine learning validations, A/B tests, and predictions
# Build models using techniques that assume normality (like Naive Bayes, linear regression)

```
